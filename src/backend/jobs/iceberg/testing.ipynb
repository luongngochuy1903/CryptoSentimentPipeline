```python
from pyspark.sql import SparkSession
spark = SparkSession.builder \
    .appName("Iceberg Init") \
    .getOrCreate()
```

```python
spark.sql("USE iceberg.default;")
spark.sql("SHOW CATALOGS").show()
```

```python
spark.sql("""
        CREATE TABLE IF NOT EXISTS iceberg.default.news (
            domain STRING,
            title STRING,
            url STRING,
            text STRING,
            published TIMESTAMP,
            author STRING,
            source STRING,
            tag STRING,
            year INT,
            month INT,
            day INT
        )
        USING ICEBERG
        PARTITIONED BY (year, month, day)
        LOCATION 's3a://your-bucket/silver/news/'
    """)
```
